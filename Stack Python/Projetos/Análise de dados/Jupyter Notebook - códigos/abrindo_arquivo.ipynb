# Mexendo com matplotlib

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
plt.suptitle("Nota redação x Escolaridade x Estado", fontsize=20)

microdados_socio_economicos.filter(items=["SG_UF_RESIDENCIA", "NU_NOTA_REDACAO", "Q001"])\
.where(microdados_socio_economicos.Q001 != "H")\
.groupby(["Q001", "SG_UF_RESIDENCIA"])\
.mean().sort_values(by="NU_NOTA_REDACAO", ascending=False)\
.unstack().plot(ax=ax, colormap="viridis", grid="True")


#mudar conteúdo da legenda
import re
handles, labels = ax.get_legend_handles_labels()
edited_labels = [re.search(",\s(.+?)\)", label).group(1) for label in labels]

#mudar posição da legenda
ax.legend(edited_labels, bbox_to_anchor=(1.05, 1), loc=0, borderaxespad=0) 

#adicionar campo de legenda formatado
textdic = ""
for k, v in dic_q.items():
    textdic = textdic + f"{k}: {v}\n"
ax.text(0.02, 0.65, textdic, transform=ax.transAxes, fontsize=14, bbox={"boxstyle": "round", "facecolor": "wheat", "alpha": 0.5})
```
