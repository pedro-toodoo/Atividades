```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
municipios = pd.read_csv("municipios.csv")
media_municipios = pd.read_csv("media_notas_por_municipio.csv", sep=",", index_col=0)
```

```python
municipios.head()
```

```python
media_municipios.head()
```

```python
enem_idhm_municipios = pd.merge(media_municipios, municipios, how="left", left_on=['CO_MUNICIPIO_RESIDENCIA'], 
                                right_on=['codigo_ibge'])
```

```python
enem_idhm_municipios.head()
```

```python
enem_idhm_municipios.plot.scatter(x="longitude", y='latitude', 
                                  c=enem_idhm_municipios['NU_NOTA_REDACAO'], 
                                  cmap='RdYlGn',
                                  figsize=(10, 8),
                                 title='NU_NOTA_REDACAO')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(20,16))

enem_idhm_municipios.plot.scatter(x="longitude", y='latitude', 
                                  c=enem_idhm_municipios['NU_NOTA_MT'], 
                                  cmap='RdYlGn',
                                  ax=ax[0,0],
                                 title='NU_NOTA_MT')

enem_idhm_municipios.plot.scatter(x="longitude", y='latitude', 
                                  c=enem_idhm_municipios['NU_NOTA_CN'], 
                                  cmap='RdYlGn',
                                  ax=ax[0,1],
                                 title='NU_NOTA_CN')

enem_idhm_municipios.plot.scatter(x="longitude", y='latitude', 
                                  c=enem_idhm_municipios['NU_NOTA_CH'], 
                                  cmap='RdYlGn',
                                  ax=ax[1,0],
                                 title='NU_NOTA_CH')

enem_idhm_municipios.plot.scatter(x="longitude", y='latitude', 
                                  c=enem_idhm_municipios['NU_NOTA_REDACAO'], 
                                  cmap='RdYlGn',
                                  ax=ax[1,1],
                                 title='NU_NOTA_REDACAO')
```
